```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv('Billionaires Statistics Dataset.csv')
heatmap_data = df.pivot_table(index='country',
                            columns='industries',
                            values='finalWorth',
                            aggfunc='count',
                            fill_value=0)
df

top_countries=df['country'].value_counts().head(10).index
top_industries=df['industries'].value_counts().head(8).index
heatmap_data=heatmap_data.loc[top_countries,top_industries]
                            
plt.figure(figsize=(12,8))
sns.heatmap(
    heatmap_data,
    cmap='YlOrBr',  # Orange-yellow color scale
    annot=True,      # Show numbers
    fmt='d',         # Format as integers
    linewidths=0.5
)
plt.title('Number of Billionaire by Country and Industry')
plt.xlabel('Industry')
plt.ylabel('Country')
plt.xticks(rotation=45,ha='right')
plt.tight_layout()
plt.show()
           
```
