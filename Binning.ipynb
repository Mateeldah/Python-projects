```python
import pandas as pd
#sample data 
df = pd.DataFrame({'age':[5,12,20,25,37,45,60,80]})
```

```python
#Define bins (edges) and labels
bins=[0,18,35,60,100]
labels = ['child', 'Young Adult', 'Adult', 'Senior']
#Apply binning
df['age_group'] =pd.cut(df['age'], bins=bins, labels=labels)
df
```

```python
df['age_quantile'] = pd.qcut(df['age'],q=4,labels=['Q1','Q2','Q3','Q4'])
df
```

```python
import numpy as np
df['bin_label'] =np.where(df['age'] <18,'Child','Adult')
df
```
