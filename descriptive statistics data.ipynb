```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('Billionaires Statistics Dataset.csv')
```

```python
df.plot.scatter(x='age',y='finalWorth',title='Age vs.Net Worth of Billionaires',alpha=0.6,figsize=(10,6),color='blue')
plt.xlabel('Age')
plt.ylabel('Net Worth (Billions $)')
plt.grid(True)
plt.show()
```

```python
wealth_trend=df.groupby('birthYear')['finalWorth'].mean()

wealth_trend.plot(kind='line',title ='Average Billionaire Wealth by Birth Year', figsize=(12,6),
color='green')
plt.xlabel('Birth Year')
plt.ylabel('Average Net Worth (Billions$)')
plt.grid(True)
plt.show()

```
