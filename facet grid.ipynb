```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv('Billionaires Statistics Dataset.csv')
df

g = sns.FacetGrid(
    df,
    col='gender',
    row='selfMade',
    height=4,
    aspect=1.5
)
g.map(sns.histplot, 'finalWorth', bins=20)
g.set_axis_labels('Net Worth (Billions $)', 'Count')
g.fig.suptitle('Wealth Distribution by Gender & Self-Made Status', y=1.05)
plt.show()
```
