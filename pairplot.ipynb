```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv('Billionaires Statistics Dataset.csv')                                                  #How to Read the Output:
                                                                       #Your pair looks like a grid with 4 variables, producing 16 subplots (4×4):

                                                                                       #1. Diagonal Plots (KDE)
                                                                                               #Show distribution of single variables

                                                                   #Example: The finalWorth KDE plot confirms most billionaires have $1B-$10B (right-skewed)

                                                                                                        #2. Off-Diagonal Scatter Plots
                                                                                              #Each shows relationship between two variables

                                                                           #Example: The age vs birthYear plot will show a perfect negative
                                                                                                      diagonal (since age = current_year - birthYear)
df

sns.pairplot(
    df[['finalWorth', 'age', 'birthYear', 'gdp_country']].sample(100),
    diag_kind='kde',
    plot_kws={'alpha':0.6}
)
plt.suptitle('Pairwise Relationships Between Key Variables', y=1.02)
plt.show()
```
