```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
df=pd.read_csv('World_population.csv')
df
```

```python
Top10=df.sort_values(by='2022 Population',ascending=False).head(10)
plt.figure(figsize=(10,5))
plt.bar(Top10['Country'],Top10['2022 Population'],color='green')
plt.title('Top 10 Most Populated Countries in 2022')
plt.xlabel('Country')
plt.ylabel('Population (2022)')
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
bar_width=0.4
x=range(len(Top10))
plt.bar([i-bar_width/2 for i in x],Top10['2000 Population'],
        width=bar_width,
        label='2000',
        color='yellow')
plt.bar([i+bar_width/2 for i in x],Top10['2022 Population'], width=bar_width, label='2022', color='teal')
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Population Comparison:2000 vs 2022 (Top 10 Countries)')
plt.grid(axis='y',linestyle='--',alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()         
```

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
df=pd.read_csv('World_population.csv')
df
```

```python
import matplotlib.pyplot as plot
df.columns=df.columns.str.strip()
top10=df.sort_values(by='2022 Population',ascending=False).head()
plt.figure(figsize=(10,6))
plt.bar(top10['Country'],top10['2022 Population']/1-000_000,color='orange')
plt.title('Top 10 Most Population Countries in 2022',fontsize=14)
plt.xlabel('Country')
plt.ylabel('Population (in millions)')
plt.xticks(rotation=45)
plt.grid(axis='y',linestyle='--',alpha=0.6)
plt.tight_layout()
plt.show()
```
